STEP-1: LOADING THE DATASET

```python
#Loading
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("trafficaccident.csv", nrows=80000)

print(data.info())
print(data.head())
```

```python
print(data.columns)
```

STEP-2: DATA CLEANING

```python
#Drop rows with missing columns
data.dropna(subset= ['CASENUMBER', 'CASEDATE', 'ACCIDENTLOC'], inplace = True)

#convert casedate to datetime
data['CASRDATE'] = pd.to_datetime(data['CASEDATE'])
```

STEP-3: EXPLORATORY DATA ANALYSIS


```python
# Convert CASEDATE to datetime
data['CASEDATE'] = pd.to_datetime(data['CASEDATE'])

# Countplot for accidents by month
data['Month'] = data['CASEDATE'].dt.month
plt.figure(figsize=(10, 6))
sns.countplot(x='Month', data=data)
plt.title("Accidents by Months")
plt.show()
```

```python
#Countplot for accidents by day of the week
data['DayOfWeek'] = data['CASEDATE'].dt.dayofweek
plt.figure(figsize=(10,6))
sns.countplot(x = 'DayOfWeek', data = data)
plt.title('Accidents by Day of the Week')
plt.show()
```
